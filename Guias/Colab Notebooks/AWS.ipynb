```python
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from osgeo import gdal
```

```python
path = 'https://landsat-pds.s3.amazonaws.com/c1/L8/009/056/LC08_L1TP_009056_20200109_20200114_01_T1/LC08_L1TP_009056_20200109_20200114_01_T1_B2.TIF'
dataset = gdal.Open(path)
b2 = dataset.GetRasterBand(1)
b2 = b2.ReadAsArray()
print(b2)
```

```python
b2=np.where(b2==0,np.nan,b2)
fig, ax = plt.subplots(1, figsize=(10, 10))
plt.imshow(b2)
plt.colorbar();
plt.show()
```
